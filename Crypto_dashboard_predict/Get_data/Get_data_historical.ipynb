```python
from binance_historical_data import BinanceDataDumper
```

# khởi tạo đối tượng BinanceDataDumper

```python
data_dumper = BinanceDataDumper(
    path_dir_where_to_dump="/Users/dophi/Desktop/LSTM_ETH/Get_data",
    asset_class="spot",  # spot, um, cm
    data_type="klines",  # aggTrades, klines, trades
    data_frequency="1d", #dữ liệu khung ngày
)
```

1. path_dir_where_to_dump:
    (string) Path to folder where to dump the data
2. asset_class:
    (string) Source of data: [spot, um, cm] um: usd(t) margined futures, cm: coin margined futures
3. data_type=”klines”:
    (string) data type to dump:
    [aggTrades, klines, trades] for spot
    [aggTrades, klines, trades, indexPriceKlines, markPriceKlines, premiumIndexKlines, metrics] for futures (metrics only supported for um)
Refer to binance doc for additional info: https://github.com/binance/binance-public-data
4. str_data_frequency:
    (string) One of [1m, 3m, 5m, 15m, 30m, 1h, 2h, 4h, 6h, 8h, 12h]
    Frequency of price-volume data candles to get

# Gọi hàm dump_data để tải dữ liệu lịch sử
1. tickers: BTC (lấy dữ liệu cho Bitcoin), ETH (lấy dữ liệu cho Ethereum)

```python
data_dumper.dump_data(
    tickers=["BTCUSDT", "ETHUSDT"],
    date_start=None,
    date_end=None,
    is_to_update_existing=False,
    tickers_to_exclude=[],
)
```

# gộp file btc thành 1 file duy nhất

```python
import pandas as pd
import glob

# đặt tên cột cho dữ liệu kline
cols = [
    "open_time", "open", "high", "low", "close", "volume",
    "close_time", "quote_asset_volume", "number_of_trades",
    "taker_buy_base_asset_volume", "taker_buy_quote_asset_volume", "ignore"
]

monthly_files = glob.glob("spot/monthly/klines/BTCUSDT/1d/*.csv")
daily_files   = glob.glob("spot/daily/klines/BTCUSDT/1d/*.csv")


files = sorted(monthly_files + daily_files)

df = pd.concat((pd.read_csv(f, names=cols) for f in files), ignore_index=True)
df = df.sort_values("open_time")
df.to_csv("/Users/dophi/Desktop/LSTM_ETH/Get_data/data/BTCUSDT_1d.csv", index=False)
```
